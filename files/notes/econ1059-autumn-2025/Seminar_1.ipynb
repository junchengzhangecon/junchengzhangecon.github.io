* Next, consider the problem from the lecture, where we calculated $1 + 2 + 3 + 4 + 5 + 6 + 7 + 8$. Write a general function that solves this problem for $n + (n+1) + (n+2) + \ldots + (n+k)$.

```python
# Arithmetic sequence
def sum_sequence(x_min, x_max):  
    result = (x_min+x_max) * (x_max-x_min+1)/2
    return result
```

```python
print(sum_sequence(1,10))
```
