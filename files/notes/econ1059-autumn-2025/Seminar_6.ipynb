## Reading Files

Use `open(path, "r")`.

*  **`read()`** → read the entire file as one string  
*  **`readlines()`** → read all lines into a list
*  **Loop line-by-line** → memory-efficient, recommended  
*  **`with`** → automatically closes the file when finished, recommended   

````python
## Example 1 — Read entire file

# My cwd is "python FALL/", if I need to open 
# "example_1.txt" inside "topic_6/" subfolder, 
# I use the relative path as below:

file_path = "topic_6/example_1.txt"

f = open(file_path, "r")
text = f.read()
print(text)
f.close()
````

````python
## Example 2 — Read all lines into a list

f = open(file_path, "r")
lines = f.readlines()
print(lines)
f.close()
````

````python
## Example 3 — Looping through a file (recommended)

f = open(file_path, "r")
for line in f:
    print(line)     # Each line ends with '\n'
f.close()
````

````python
## Example 4 — with automatically closes the file (recommended)

with open(file_path, "r") as f:
    text = f.read()

print(text)
````

`strip()` family removes characters from **string edges only**, not from the file itself. They return a **new string** → original unchanged.

| Method      | Removes from | Default removed chars |  Example | Output |
|-------------|--------------|-----------------------|---------------|--------|
| `strip()`   | both sides   | whitespace (`\n`, space,`\t`) | `"  hi  ".strip()` | `"hi"` |
| `lstrip()`  | left side    | whitespace            | `"  hi  ".lstrip()` | `"hi  "` |
| `rstrip()`  | right side   | whitespace            | `"  hi  ".rstrip()` | `"  hi"` |

````python
## Example 4 — Removing specific characters

s = "hello!!!#@$"
print(s.rstrip("!#@"))   # Starts from the RIGHT end ('$'), last char is '$' → NOT in {"!", "#", "@"}

s = "hello!!!#@$!!##@"
print(s.rstrip("!#@"))   # → "hello!!!#@$"

s =  "hel@@lo!!!#@$!!##@"
print(s.rstrip("!#@"))   # → "hel@@lo!!!#@$"

s = "@!@@$@hello!!!#@$"
print(s.lstrip("!#@"))   # → "$@hello!!!#@$"
````

## Dictionaries (`dict`)

*   Dictionaries store data in **key-value pairs**(curly brackets). You can think of a real dictionary as a collection of **pairs** where each **key** maps to a **value**.
*   Keys must be immutable (`str`, `int`, `float`, `bool`, `tuple`; **NOT** `list`, `dict`)
*   Values can be any data type
*   Dictionaries are mutable, can add, update, or delete key–value pairs

**In Python**:

`dict_1 = {key_1: value_1, key_2: value_2}`

````python
## Example 10 — Dictionary

# Create 
student = {
    "name": "Alice",
    "age": 19,
    101: "ID number"
}

# Accessing Values
print(student["name"])   # → Alice
print(student[101])      # → ID number

# Adding / Modifying
student["major"] = "Economics"   # add new pair
student["age"] = 20              # modify existing value
print(student)
````

````python
# Deleting
del student["major"]
print(student)
````

````python
# Keys / Values / Items
print(student.keys())      # dict_keys(['name', 'age', 101])
print(student.values())    # dict_values(['Alice', 20, 'ID number'])
print(student.items())     # dict_items([('name', 'Alice'), ('age', 20), (101,'ID number')])
````

````python
# Key existence
if "age" in student:
    print("Found age")
````

* Let's start with a decent, albeit slighly insufficient solution
* Essentially, the code iterates over the list of words and check if the current element of the list matches the word we are looking for
* Please note: like most of you, I first wrote (a somewhat messy) code that didn't use a function. Once it worked, I turned it into a beautiful, general function. This question surely wasn't easy.

````python
def count_words(word_to_count, text):
    count = 0
    
    text_list = text.split()
    # print(text_list)  # Insightful! Uncomment to see what happens
    
    for word in text_list:
        if word_to_count == word:
            count += 1
            
    return count
````

````python
print(count_words("they", original_text))
````

````python
print(count_words("nonsense", original_text))
````

* It turns out that the last element of the list is "nonsense." not "nonsense"
* the `split()` function doesn't get rid of characters like .,:;!?
* $\to$ we need to clean up the original text a bit more
* The easiest way to do this is via `rstrip` where any elements you want deleted need to be put into a **single** text, e.g. `rstrip(".,:;!?")`

````python
def count_words_improved(word_to_count, text):
    count = 0

    # Clean Text
    text_list = text.split()
    text_list_cleaned = []

    for word in text_list:
        text_list_cleaned.append(word.rstrip(".,:;!?"))

    # Search Text
    for word in text_list_cleaned:
        if word_to_count == word:
            count += 1
            
    return count
````

````python
print(count_words_improved("nonsense", original_text))
````
