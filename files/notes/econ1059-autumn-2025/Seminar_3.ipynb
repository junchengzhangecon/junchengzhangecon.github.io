### Lists

- A **list** is an **ordered** and **mutable** collection in Python.  
- It can store **different data types** (e.g., integers, strings, floats), with elements separated by commas.

#### Common List Operations & Methods

| Method/Operations | Description |
|---------|------------|
| `list.append(x)` | Add `x` to the **end** of the list |
| `list.extend(iterable)` | Add **all elements** from another iterable to the end |
| `list.pop(index)` | Remove and **return** the element at `index` (default: last item) |
| `list.remove(x)` | Remove the **first occurrence** of value `x` |
| `list.index(x)` | Return the **first index** where `x` appears |
| `list.count(x)` | Return how many times `x` appears in the list |
| `list[a:b]` | Slice — get elements from index `a` to `b-1` (left-closed, right-open) |
| `len(list)`| Returns the **number of elements** in the list.

````python
def print_lazy_triangle(dim, char):
    # Outer loop: controls the number of rows to print
    for row in range(dim):
        
        # Start each row as an empty string
        # We will "grow" this string by adding characters in the inner loop
        row_to_print = ""
        
        # Inner loop: controls how many characters we add to the current row
        # Note: using (1 + row) ensures the triangle grows as 1, 2, 3, ...
        # If we omit the "1 +", the first row would be empty (0 characters)
        for column in range(1 + row): 
            # Add one character at a time
            row_to_print += char  
        
        # After finishing the inner loop, print the entire row
        print(row_to_print)

print_lazy_triangle(4, "#")

#  Out:
#  #    (Row 0, range(1) -> contains 1 character "#")
#  ##   (Row 1, range(2) -> contains 2 characters "#")
#  ###  (Row 2, range(3) -> contains 3 characters "#")
#  #### (Row 3, range(4) -> contains 4 characters "#")
````

````python
def print_christmas_tree(height, char):
    # Note: Height determines rows of needles. 
    # Height must be 2 or larger.

    if height > 1:
        # Build  the 'needle part'
        for row in range(height): 
            background_left = ""
            foreground = ""
    
            # The easiest way to create the 'text' is multipliction.
            # For loops work, but I find them a bit more complicated.
            background_left = (height - row - 1)*" "  # Empty space.
            foreground = (1 + 2*row)*char
            print(background_left + foreground)
        
        # Add stem
        stem = (height - 1)*" " + char
        print(stem)
    else:
        print("Error, a Christmas tree must have at least a height of 2")


print_christmas_tree(5, "*")
````

### Christmas Tree — Row Breakdown (Example: `height = 5`, `char = "*"`)  

| Row | `background_left` (spaces) | `foreground` (stars) | Printed Result |
|------|---------------------------|-----------------------|----------------|
| 0 | `(5 - 0 - 1) = 4` spaces | `1 + 2*0 = 1` star | `"    *"` |
| 1 | `(5 - 1 - 1) = 3` spaces | `1 + 2*1 = 3` stars | `"   ***"` |
| 2 | `(5 - 2 - 1) = 2` spaces | `1 + 2*2 = 5` stars | `"  *****"` |
| 3 | `(5 - 3 - 1) = 1` space  | `1 + 2*3 = 7` stars | `" *******"` |
| 4 | `(5 - 4 - 1) = 0` spaces | `1 + 2*4 = 9` stars | `"*********"` |
| Stem | `(5 - 1) = 4` spaces | `1` star | `"    *"` |

**Explanation:**

- `background_left` controls **how many leading spaces** we print to center the tree.
- `foreground` controls **how many characters (`char`)** appear in that row.
- Each row increases by **2 characters**, forming a symmetric triangle.

## Editing Lists
* Suppose you have a list of simple numbers collected from answers to survey questions. For example, you may have asked people "how many job applications do you plan to make daily". It turns out that some people actually claimed crazy plans such as 10 000 or 500. (that did not match their subsequent real life choices)
* Using a simple example list that you create, write an algorithm that creates a copy of this list, but where extreme values are replaced with something sensible like 20.
* Aftwerwards, turn this algorithm into a general function

**Hints and further comments for this question**
* don't manipulate the original list (bad practise). Instead store values to a new list 
* For real data analysis, it is usually better to treat items as 'missing' (and omit them from analysis) instead of arbitrarily changing them. One way to do that would be to set it to -1, if all other numbers are positive, and omitt negative values from the analysis

````python
def censor_values(list_values, max_value_allowed):
    # Replaces any item in list_values that 
    # exceeds the maximum value allowed by the 
    # maximum value allowed and returns the new list.
    
    new_list = []
    for value in list_values:
        if value <= max_value_allowed:
            new_list.append(value)
        else:
            new_list.append(max_value_allowed)
    
    return new_list

my_list = [1, 100, 10000, 500, 4]
print(censor_values(my_list, 200))
````
