## Tuples

- A **tuple** is an **ordered** and **immutable** collection of elements.
- Created using **parentheses** `()`, and can store **different data types** (e.g., `("world", 2)`).
- For a **single element**, write `(2,)` with a trailing comma — `(2)` is just an integer, not a tuple.

**Why Use Tuples?**
- **Prevent unintended modification**: Use a tuple when the data should not change.
- **Multiple return values**: Functions that return more than one value automatically return them as a **tuple**, which can then be unpacked (e.g., `x, y = func()`).



```python
# Example 1: Returning multiple values as a tuple
def function_name(param1, param2):
    add = param1 + param2
    multiply = param1 * param2
    return add, multiply   # returns a tuple (add, multiply)

results = function_name(1, 2)
print(results)      
```

```python
# Tuples are immutable 
results[0] = 8       #  TypeError
```

```python
# Example 2: Unpacking (assign the value)
sum_result, product_result = function_name(3, 4)

print(sum_result)       # Output: 7
print(product_result)   # Output: 12
```


* Create any tuple that you like, access one of its values.

## `zip()`

- `zip()` is a built-in function that **combines two or more sequences** (e.g., lists, tuples, **strings**, range) into **tuples**. 
- Elements are paired **by their positions** (1st with 1st, 2nd with 2nd, etc.).
- It behaves like a **one-time parcel**: once unpacked (e.g., converted to a list or looped over), it becomes empty and must be recreated to use again.


```python
## Example 3: Basic usage of zip()
names = ["Alice", "Bob", "Charlie"]
scores = [85, 92, 78]

combined = zip(names, scores) 

print(combined)             # A lazy zip object (iterator) — like a parcel: you must open it to see what's inside
print(type(combined))       # Notice this is neither a list nor a tuple
print(list(combined))       # Convert to a list → opens the parcel and consumes the iterator

# Possible Output:
# <zip object at 0x000001A3B8F...>
# <class 'zip'>
# [('Alice', 85), ('Bob', 92), ('Charlie', 78)]
```

```python
print(list(combined))       # Empty — the parcel has been opened; re-zip to get items again
```

```python
## Example 4: zip() stops at the shortest sequence
names = ["Alice", "Bob", "Charlie"]
scores = [90, 85]

for name, score in zip(names, scores):   # Unpacking
    print(name, score)   
# Output:
# Alice 90
# Bob 85
```

```python
## Example 5: Using strict=True
names = ["Alice", "Bob", "Charlie"]
scores = [90, 85]

for name, score in zip(names, scores, strict=True):
    print(name, score)

# Output:
# ValueError: zip() argument 2 is longer than argument 1
```

```python
## Example 6: It also works for strings and range! String is a sequence of characters after all. 
word = "cat"
for ch in word:
    print(ch)
for ch, pos in zip(word, range(3)):
    print(ch, pos)
```


* Use the zip() function to create a list of tuples of the form (1,10), (2, 11), (3, 12), (4, 13).


```python
# Basic Solution
new_list = []

for item in zip(range(1, 5), range(10, 14)): 
    print(item)
    new_list.append(item)
    
print(new_list)
```

## `enumerate()`

- `enumerate()` is a built-in function that adds an automatic **index counter** to an iterable.
- It returns **(index, element)** pairs, allowing you to access both the position and the value in a loop.

```python
## Example 6

fruits = ["apple", "banana", "cherry"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
    
```

## Docstring

A **docstring** (documentation string) is a string literal used to describe a Python module, function, class, or method.  
It is written inside **triple quotes `"""`** and placed **immediately after the definition**.


### Writing Docstrings

- Enclosed in triple quotes, e.g., `"""My docstring"""` (single-line example).
- Multi-line docstrings should end with `"""` on a **new line**.
- Followed by a **blank line**.
- Should not repeat the name of the function/class — instead, explain **what it does** and **what it returns**.


#### Google Style Guide

- Recommends adding a docstring to any function, method, etc., that meets at least one of the following:
  - Part of the public API
  - Non-trivial in size
  - Contains non-obvious logic

- A well-structured docstring typically includes:
  * General description — What the function does
  * Arguments — The parameters it takes
  * Returns — What it outputs
  * Raises (exceptions) — Possible errors


```python
## Example 7 Google Docstring Style
def add_numbers(a, b):
    """Add two numbers together.

    Args:
        a (int or float): First number.
        b (int or float): Second number.

    Returns:
        int or float: The sum of a and b.

    Raises:
        TypeError: If either input is not numeric.
    """
    return a + b
```

```python
# Get the docstring
help(add_numbers)
```

```python
# Get the docstring
print(add_numbers.__doc__)
```

```python
# And with the long - and indeed much more helpful - docstring:

def find_first_instance_in_list(li, value_to_find):
    """Finds the index of the first occurrence of the value in a list.

    Args:
        li: The list to be searched.
        value_to_find: The value to locate within the list.

    Returns:
        The index of the first occurrence of the value, or None if the 
        value is not found.
    """
    
    for i, val in enumerate(li):
        if val == value_to_find:
            return i
            
    return None   
```

**General Algorithm of Hangman**
* [1] A secret word is selected, and numbers of letters are indicated as follows:
    * "_ _ _ _ _ _"
* [2] Guessing Starts, which has the following structure:

* [3] If 7 misses: GameOver, otherwise
    * [3.1] guess a letter
    * [3.2]check if letter is part of secret word;
       * if true, write letter at the correct place of  "_ _ _ _ _ _"
       * if false, increase misses by 1; 
           * advanced features: track wrong letters
           * advanced features: draw next part of the gallows
    * go back to [3].

```python
secret_word = "hello"  # Set word to test out our functions

# Create a list of the same length as our secret word, that contains "False" for each element
# as initially, no letter has been guessed correctly.
guessed_correctly = len(secret_word)*[False]  
past_guesses = []  # To track guesses
total_misses_allowed = 6  # Do not change
total_misses = 0  # Tracks incorrect answers                       

while total_misses < total_misses_allowed:                          
    # A. Print secret word
    to_print = ""    
    for letter, guess in zip(secret_word, guessed_correctly):       
        if guess == True:
            to_print += letter + " " 
        else:
            to_print += "_ "         
    print(to_print)

    # B. Print guesses already made; simplest possible implementation
    print(f"Already guessed: {past_guesses}.\n") 

    # C. Elicit new guess
    guess = input(f"Please make a guess ({total_misses_allowed - total_misses} mistakes left):\n")   

    # D. Check guess    
    guess_is_correct = False

    for i, secret_word_char in enumerate(secret_word):              
        if guess == secret_word_char:
            guess_is_correct = True
            guessed_correctly[i] = True                                              
            # Not breaking the loop since the same letter could show up multiple times.
            # I also don't care if we check a position of the word that has already been guessed.
            # Notice, this also means you can guess the same correct letter multiple times.

    # E. Print outcome of guess; adjust misses and past guesses
    if guess_is_correct:                                            
        print("Correct!\n\n\n")
    else:    
        total_misses += 1
        past_guesses.append(guess)  # I opted for only tracking misses.
        print(f"Incorrect!\n\n\n")
        
    # F. Check if game won/ game over.
    if guessed_correctly == len(secret_word)*[True]:  # Feels very ugly; but works.    
        print("!!! You win !!!")
        break
    if total_misses == total_misses_allowed:
        print("--- Game Over ---")
        break

    # initialization  - main body (iterations) - endding conditions
    # Intended block!!! 
```
