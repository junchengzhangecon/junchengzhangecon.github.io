###  Booleans

- A **Boolean** value represents **truth** — it can be either `True` or `False`.  
- Any **comparison expression** (such as `x == y`, `x > y`, `x >= y`, or `x != y`) returns a Boolean result.

| Operator | Meaning | Example | Result |
|:---------:|:--------|:--------|:--------|
| `==` | Equal to | `5 == 5` | `True` |
|   `!=` | Not equal to | `5 != 3` | `True` |
| `>` | Greater than | `7 > 10` | `False` |
| `<` | Less than | `3 < 8` | `True` |
| `>=` | Greater than or equal to | `10 >= 10` | `True` |
| `<=` | Less than or equal to | `4 <= 2` | `False` |   

###  Combining Multiple Booleans

You can combine or modify Boolean values and comparison results using **logical operators**:

- **`and`** → Returns `True` **only if both** statements are `True`.  
- **`or`** → Returns `True` **if at least one** statement is `True`.  
- **`not`** → **Negates** a Boolean value (changes `True` to `False`, and vice versa).

````python
## Example 1

print(True and False)# False  because both must be True for 'and' to be True
print(True or False) # True   because at least one is True for 'or' to be True
print(not True)      # False  'not' reverses the value
print(not False)     # True   'not' reverses the value
````

````python
## Example 2

x = 10
y = 5

print(x == 10 and y != 10)   # True (x == 10 is True, y != 10 is True)
print(x > 5 or y > 10)       # True (x > 5 is True)
print(not(x < 5))            # True (x < 5 is False, but negated)
````

###  `if/elif/else`

- Used for **decision-making** in Python.

- Allows the program to **choose different actions** based on different conditions.

- Basic structure:

```python
    if condition1:        # Check the first condition (colon ':')
        do_something_1    # Runs if condition1 is True (indented block) 
    elif condition2:      # Checked only if condition1 is False
        do_something_2    # Runs if condition2 is True
    else:                 # Runs only if all above conditions are False
        do_something_else
```

  

  - `if` checks the **first condition**.

  - `elif` checks **another condition** if the previous `if` was False. (You can have multiple `elif` blocks.)

  - `else` runs **when all previous conditions are False**.

````python
## Example 3

x = 10

if x > 15:
    print("x is greater than 15")
elif x > 5:
    print("x is greater than 5 but not greater than 15")
else:
    print("x is 5 or less")
````

* Write a function that takes two arguments. It should subtract the second argument from the first if both arguments are positive and returns the solution. It returns 0 if one or both arguments are negative.

````python
def subtract_two_positive_numbers(x, y):
    if x >= 0 and y >= 0:
        return x - y
    else:
        return 0
````

````python
# Let's test it
print("4-1 yields", subtract_two_positive_numbers(4, 1))
print("4-(-10) yields", subtract_two_positive_numbers(4, -10))
````

````python
def subtract_two_positive_numbers_v2(x, y):
    result = 0
    if x >= 0 and y >= 0:
        result = x - y
    return result
````

###  `range()` Function

The **`range()`** function in Python is used to **generate a sequence of numbers** — often for use in loops, especially `for` loops.

```python
range(start, stop, step)

| Parameter | Description | Default |
|:----------:|:-------------|:--------:|
| `start` | First number in the sequence (included) | `0` |
| `stop` | Number where the sequence ends (not included) | — |
| `step` | Difference between each number (increment) | `1` |

````python
## Example 7
print(list(range(5)))          # [0, 1, 2, 3, 4]  list() makes a list — a collection of items in order.
print(list(range(1, 6)))       # [1, 2, 3, 4, 5]
print(list(range(0, 10, 2)))   # [0, 2, 4, 6, 8]
print(list(range(10, 0, -2)))  # [10, 8, 6, 4, 2]
````

### `while` Loop

A **`while` loop** repeats a block of code **as long as a given condition is `True`**.  
It is often used when you **don’t know in advance** how many times the loop should run.
```python
while condition:
    # code block to repeat
    # (make sure to update something inside the loop)
```
- The condition is checked **before each iteration**.  
- The code inside the loop runs **only if** the condition is `True`.  
- The condition must be **updated inside the loop**, otherwise it will cause an **infinite loop**.

- **`break`** → Stops the loop **immediately**, even if the condition is still `True`.

- **`continue`** → Skips the rest of the **current iteration** and moves on to the next one.

- **`else`** → Runs a block of code **once after** the loop ends normally (when the condition becomes `False` or the loop finishes without `break`).



````python
## Example 8

count = 0
    
while count < 5:
    print(count)
    count += 1
````

````python
## Example 9

i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
````

````python
## Example 10
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
````

````python
## Example 11
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print("i is no longer less than 6")
````

| Aspect | `for` Loop | `while` Loop |
|:--|:--|:--|
| **Usage** | For **known** number of iterations | For **unknown** repetitions until a condition fails |
| **Control** | Iterates over a **sequence** | Runs **while a condition is True** |
| **Setup** | Counter defined **in the loop** | Variable defined **before** the loop |
| **Risk** | Rarely infinite | Can be **infinite** if condition never changes |

* `for` Loop to check whether a coin is fair 

*(We haven’t learned all the code here yet — for now, just know that `for` loops are very useful in simulations, so don’t worry about the details!).*

````python
import random

heads = 0
tails = 0

# Flip the coin () times
for i in range(10000):
    # random.choice() selects one element from the list uniformly at random.
    # Here, "Heads" and "Tails" each have probability 0.5. 
    flip = random.choice(["Heads", "Tails"])
    if flip == "Heads":
        heads += 1
    else:
        tails += 1

# Print the counts
print("Heads:", heads)
print("Tails:", tails)
````
