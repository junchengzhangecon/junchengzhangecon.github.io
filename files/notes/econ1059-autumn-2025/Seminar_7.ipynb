## Classes


A class puts together:

*   data (stored in each instance)

*   methods (functions acting on that data)

This lets you build many objects with similar structure but different values (e.g., different firms or consumers), helping code stay clean and organized.

**Basic structure**:

```python

class ClassName:
    """
    Optional: Class docstring
    Describe what this class represents and what it is used for.
    """

    # Constructor
    # Initializes instance attributes
    def __init__(self, arg1, arg2, ...):
        self.attr1 = arg1      # instance attribute
        self.attr2 = arg2
        # ...

    # Instance method
    # Behavior acting on instance data
    def method1(self, ...):
        # access instance data via self
        # do something
        return ...
    
    # Optional: String representation
    # What print(obj) shows
    def __str__(self):
        return f"ClassName(attr1={self.attr1}, attr2={self.attr2}, ...)"

### Inheritance
Inheritance allows a class (child/subclass) to reuse and extend the properties and methods of another class (parent/superclass).

*   Parent → Child structure (The child automatically gets attributes(methods) defined by the parent)
```python
class Parent:
    ...

class Child(Parent):
    ...
```

*   Reuse parent initialization: **`super()`**
```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

class Square(Rectangle):
    def __init__(self, length):
        super().__init__(length, length)   # call parent __init__
```
*   Child can add new features (even override)


````python
# Example 2 Dog class Inheriting from Pet (Part of ans to Q1)
class Dog(Pet):
    """
    Dog class inherits from Pet.
    Adds breed information.
    """
    def __init__(self, name, breed, age, gender):
        # Call parent constructor, species always 'dog'
        super().__init__(name, age, gender, species="dog")
        self.breed = breed
    
    # Override introduce to include breed
    def introduce(self):
        return (
            f"My name is {self.name}. "
            f"I am a {self.age}-year-old {self.gender} {self.breed}."
        )
    
    def bark(self):
        return f"{self.name} says: Woof! Woof!"


marcus_dog = Dog("Rhuby", "von Rhubarb", 1, "female")
my_dog     = Dog("Siri", "Shiba Inu", 4, "male")

print(marcus_dog.introduce())
print(marcus_dog.bark())
print("")
print(my_dog.introduce())
print(my_dog.bark())
````

## Playing with Rectangles
* Create a rectangle class that includes a method that cuts the rectangle in half (along the largest side)
    * Can you write the cut-method so that it returns the cut-away part of the rectangle as an object iself
    * Finally, write a function (using the above class) to calculate the sum of the areas of the following objects: (the area of) an initial square, (the area of) the square cut in half, (the area) of the cut-in-half square cut in half, ..., up to n-cuts! Needless to say, this function should use your class-objects (not your advanced understanding of math).
* **Note on how to tackle this question:** while the above question text guides you towards (what I think) is a very cool implementations, there are other solutions that can solve this problem just as well. If you find the above hard to implement, try a different approach.

````python
class RectangleCutable:  
    """ A Rectangle that can be cut in half along the longest edge.
    
    Attributes: 
        width: the width of the rectangle
        height: the height of the rectangle
    """
    
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def get_area(self):
        return self.width * self.height

    def cut_in_half(self):
        """ Cut the current rectangle in half
            return: cut-away part of rectangle as a new Object.
        """
        
        if self.width >= self.height:
            self.width = self.width / 2
        else:
            self.height = self.height / 2
        
        return RectangleCutable(self.width, self.height)    

    def __str__(self):
        return f"Rectangle of width = {self.width} and height = {self.height}."
````

````python
# Testing RectangleCutable
rectangle_0 = RectangleCutable(6, 8)
print(f"Original rectangle: {rectangle_0}")
````

````python
rectangle_1 = rectangle_0.cut_in_half()
print(f"New Rectangle, after one round of cutting:\n {rectangle_1}")
````

````python
print(f"Old Rectangle, after one round of cutting:\n {rectangle_0}")
````

### Solution to:
* Calculate the area of the series of n rectangles where the next rectangle is half of the previous.

````python
def find_area_series_cut_rectangles(initial_height, initial_width, num_cuts):
    rectangle = RectangleCutable(initial_height, initial_width)
    area = rectangle.get_area()
    for i in range(num_cuts):
        rectangle = rectangle.cut_in_half()
        area += rectangle.get_area()
    return area
````

````python
# Test find_area_series_cut_rectangles
height = 1
width = 1
print(f"Area of a series of rectangles, with the first being a 1x1 square, the scond a 0.5x1 rectangle is:")
print(find_area_series_cut_rectangles(height, width, 1))

print(f"For: 1x1, 0.5x1, and 0.5x0.5, we have:")
print(find_area_series_cut_rectangles(height, width, 3))
````

### `@staticmethod`

A static method is a function defined inside a class that:

*   does not take `self`

*   does not use instance attributes

It behaves like a **normal function**, but is grouped inside the class for organization.

````python
## Example 3: Dog class with static method 
class Dog(Pet):
    """
    Dog class inherits from Pet.
    Adds breed information and dog-specific behavior.
    """

    # Constructor
    # name, breed, age, gender → passed in
    def __init__(self, name, breed, age, gender):
        # Call parent(Pet)'s constructor
        super().__init__(name, age, gender, species="dog")     
        self.breed = breed   # new attribute only in Dog
    
    # Instance method: needs self
    def bark(self):
        return "Woof! Woof!"

    # Static method: does NOT take self, universal dog sound
    @staticmethod
    def bark_static():
        return "Woof! Woof!"
````

````python
# Create a Dog instance
my_dog = Dog("Siri", "Shiba Inu", 4, "male")

# When calling introduce():
# → Python runs: Pet.introduce(my_dog) → self = my_dog
print(my_dog.introduce())   # ← inherited from Pet
print(my_dog.bark())        # ← Dog instance method, we know who is barking - Siri
print(my_dog)               # __str__ from parent Pet

print("")
# Static method behaves like a normal function placed inside the class
print(Dog.bark_static())    # ← static method, we don't know who is barking, just returns a sound
````

## Inheritance
* In the class `square` of our lecture notes, we called `super().__init__(length, length)` in the `__init__` method. How could you write the class without relying on the parent's initialization?
<br>
(If you get it right on the first try, also try to write a version that doesn't work to make sure whether you really understand it).

* Shortest solution (all pointless stuff cut)

````python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def get_area(self):
        return self.width * self.height


class Square(Rectangle):
    def __init__(self, length): #override the initializer
        # super().__init__(length, length)
        self.width = length
        self.height = length
    
    def get_diagonal(self):
        return (2*self.width**2)**0.5
````

````python
my_square = Square(4)
print(f"Area = {my_square.get_area()}")
````

````python
my_broken_square = SquareBroken(4)
print(f"Diagonal = {my_square.get_diagonal()}")
````
