- **`import random`**  
  Imports the entire module. Call functions with a prefix: **`random.randint(1,10)`**.

#### Random Module Summary

- **`random.choice(sequence)`**  
  Selects **one random element** from a **non-empty sequence** (e.g., list, tuple, string).

- **`random.random()`**  
  Returns a random **float** in the range **[0.0, 1.0)** (includes `0.0` but **excludes** `1.0`).

- **`random.uniform(a, b)`**  
  Returns a random **float** in the range **[a, b]** (both endpoints **may be included**, and works even if `a > b`).

- **`random.randint(a, b)`**  
  Returns a random **integer** in the **closed interval [a, b]** (both `a` and `b` are included).

- **`random.randrange(start, stop, step)`**  
  Returns a random **integer** from a range, including `start` but **excluding** `stop`.

- **`random.gauss(mu, sigma)`**  
  Returns a random float drawn from a **normal (Gaussian) distribution**  
  with **mean = `mu`** and **standard deviation = `sigma`**.

- **`random.seed(value)`**  
  Sets the **seed** for the random number generator to ensure **reproducible results**.  
  In Jupyter Notebook, the seed should be placed in the **same cell** as the random function.  
  Without setting a seed, random results will **change on each run**.

### `matplotlib` 

- **`import matplotlib.pyplot as plt`**  
  Imports the plotting interface.

- **`fig, ax = plt.subplots()`**  
  Creates a Figure (**`fig`**) containing one Axes (**`ax`**).

- **Common plotting methods (called on `ax`)**  
  - Line: **`ax.plot(x, y)`**  
  - Scatter: **`ax.scatter(x, y)`**  
  - Bar: **`ax.bar(x, height)`**  
  - Pie: **`ax.pie(sizes, labels=..., autopct=...)`**

- **Labels, title, legend**  
  - **`ax.set_xlabel("text")`**  
  - **`ax.set_ylabel("text")`**  
  - **`ax.set_title("text")`**  
  - **`ax.legend()`**  

- **Layout & display**  
  - **`plt.tight_layout()`** — avoids overlapping when multiple subplots exist  
  - **`plt.show()`** — displays the final figure

```python
## Example 4 Simple Line
import matplotlib.pyplot as plt

# Create figure and axes 
# figsize=(3, 2) means 3 inches wide × 2 inches tall. Default size(6.4, 4.8)
fig, ax = plt.subplots(figsize=(3, 2)) 

# Data
x = [1, 2, 3]
y = [2, 4, 6]

# Draw a line on this axes
ax.plot(x, y)

# Add labels
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Simple Line")

# Show figure
plt.show()
```

* Create a scatterplot graph for data generated by $y = x/2 - 2 + e$, where $e$ is a random error that follows a uniform distribution $U[-1, 1]$. Your graph should include 30 datapoints on the interval $[0, 10]$.

**Comments on Solution**
* As the question does not say anything about how $x$ is generated, it makes sense to also generate it randomly (but having 30 equally spaced datapoints is ok too).
* Aside: the following code is a great starting point to generate data for regressions (feel free to replace the uniform error with a normally distributed error). You can even use it to create hypothetical data to pre-test the analysis of an upcoming experiment that you designed (to make sure your experiment is well designed, and your analysis code works)

```python
import matplotlib.pyplot as plt
import random

random.seed(1234)
data_x = []
data_y = []

for i in range(30):
    #x = i * (10 / 29)            # evenly spaced between 0 and 10
    x = random.uniform(0, 10)
    y = x/2 - 2 + random.uniform(-1, 1)
    data_x.append(x)
    data_y.append(y)
```

```python
fig, ax = plt.subplots()
ax.scatter(data_x, data_y)
plt.show()
```

- **`random.gauss(mu, sigma)`**

```python
## Example 5 200-Point Scatter: x ∈ [−5, 5], y ~ N(0, 1)
import matplotlib.pyplot as plt
import random

# Start from empty lists
x_values = []
y_values = []

# Generate 200 points
for i in range(200):
    # even spacing from -5 to 5
    x = -5 + i * (10 / 199)        # 10 = width of interval; 199 = steps
    y = random.gauss(0, 1)         # Gaussian noise N(0,1)

    x_values.append(x)
    y_values.append(y)

# Create figure and axis
fig, ax = plt.subplots()

# Scatter plot
ax.scatter(x_values, y_values)

# Labels
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Scatter: Gaussian Noise (mean=0, std=1)")

# Show
plt.show()
```

```python
## Example 6 Coin Flip Counts (Bar chart & Pie chart)
import random
import matplotlib.pyplot as plt

heads = 0
tails = 0

# Flip the coin 10,000 times
for i in range(10000):
    # random.choice() selects one element uniformly at random
    flip = random.choice(["Heads", "Tails"])
    if flip == "Heads":
        heads += 1
    else:
        tails += 1

# Print the counts
print("Heads:", heads)
print("Tails:", tails)
```

```python
# Data
labels = ["Heads", "Tails"]
values = [heads, tails]

# Create plot (bar)
fig, ax = plt.subplots()
ax.bar(labels, values)

# Labels
ax.set_ylabel("Count")
ax.set_title("Coin Flip Results(Bar)")

plt.show()
```

```python
# Create plot (pie)
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct="%1.1f%%")

ax.set_title("Coin Flip Results (Pie)")
plt.show()
```

- **`fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,4))`**  
  Creates a Figure (**`fig`**) with a grid of Axes (**`ax`**).  
  - **`nrows`**: number of rows (default `1`)  
  - **`ncols`**: number of columns (default `1`)  
  - **`figsize`**: figure size in inches (default ≈ `(6.4, 4.8)`)  
  Example:  
  - `plt.subplots(1,2)` → 1 row, 2 columns (2 subplots)  
  - `plt.subplots(2,2)` → 2×2 grid (4 subplots)


```python
## Example 7 Combine

# Data 
labels = ["Heads", "Tails"]
values = [heads, tails]

# Create a figure with 1 row, 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Bar chart
ax1.bar(labels, values)
ax1.set_title("Bar Chart")
ax1.set_ylabel("Count")

# Pie chart
ax2.pie(values, labels=labels, autopct="%1.1f%%")
ax2.set_title("Pie Chart")

plt.tight_layout()
plt.show()
```

## Simulate Random Events with random()
* Coinflipping. Flip a 'coin' n-times using the random integer function. What is the frequency (percentage) of heads in your sample?

```python
# Solution 1: most basic implementation.
import random

def calculate_heads_n_coinflips(num_coinflips):
    num_heads = 0 
    for i in range(num_coinflips):
        x = random.randint(0,1)
        num_heads = num_heads + x
        
    avg = num_heads / num_coinflips
    return avg
    
print(f"{calculate_heads_n_coinflips(10000): 0.3f}")
```

```python
# Solution 2: tracks all coinflips before calculating average. 
import random  # To keep examples self-contained, I import again.

def flip_coin_n_times(num_coinflips):    
    data = []
    for i in range(num_coinflips):
        data.append(random.randint(0,1))
    return data

total_flips = 10000
coinflips = flip_coin_n_times(total_flips)

# Using build-in function sum(). Could you write it yourself?
avg_heads = sum(coinflips)/total_flips  
print(avg_heads)
```

* Neither Solution 1 or Solution 2 is inherently better than the other
* Solution 1 is probably more to the point of the question.
* Solution 2 is a more general approach to these kind of questions, which could be useful in other settings or to solve more general problems. However, it is less memory efficient

## "Guess a Number" Game
* Create a game where your aim is to guess a number (among 1,...,10) that the computer secretly selected (at random) within 3 tries. 
* Note: recall that input() returns a string. Turn it into an integer with int().
* Make sure to test your game properly! In this case, you may either want to print() the secret number or get rid of the randomness for a while

### Comments on Solution
* The key idea in the code below is to have a while loop that repeats the relevant if-clauses until the game is over.
* The official solution relies on a very simple implementation, and is far from ideal. There is too much code-repetition. Maybe you can try and improve it (in a way that is still easily readable)
* Another nice extension you could add is to track previous guesses
* Jupyter notebook will run this code until you win/lose. To stop early, simply interrupt Kernel and start over
* Solution may be difficult to read in html-slides. Please see pdf or original code in that case.

```python
import random

def guess_a_number_game():
    total_misses_allowed = 3  # Set "difficulty"
    secret_number = random.randint(1,10) 

    total_misses = 0  # Do not set.

    while total_misses < total_misses_allowed:  
        # The strict inequality is correct. Try to see for yourself

        print(f"You have {total_misses_allowed - total_misses} more guesses).") 
        guess = int(input(f"Please make a guess."))
        
        if guess == secret_number:
            print("\nWah, you're a genius. You won!")
            break # Not ideal. Maybe you can write this better
        elif total_misses < total_misses_allowed - 1:
            print(f"\nNope, {guess} is not correct.\n")
            total_misses += 1
        else:
            print(f"\nNope, {guess} is not correct. \n")
            print("GameOver")
            total_misses += 1

guess_a_number_game()  # Run game 
```
